Q1. Explain the concept of homogeneity and completeness in clustering evaluation. How are they calculated?

Ans1- Homogeneity and completeness are two evaluation metrics used in clustering. Homogeneity measures how pure a cluster is with respect to a given ground truth, and completeness measures how well all data points that belong to the same class are assigned to the same cluster. Both metrics are calculated based on the contingency table, which summarizes the correspondence between the true labels and the predicted labels. Homogeneity is calculated as the ratio of the number of data points in the largest cluster that belong to the same class, to the total number of data points in that cluster. Completeness is calculated as the ratio of the number of data points in the same class that belong to the largest cluster, to the total number of data points in that class.
