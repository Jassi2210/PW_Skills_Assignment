Q1. Pearson correlation coefficient is a measure of the linear relationship between two variables. Suppose
you have collected data on the amount of time students spend studying for an exam and their final exam
scores. Calculate the Pearson correlation coefficient between these two variables and interpret the result.

Ans1. To calculate the Pearson correlation coefficient between two variables, you first need to calculate the covariance and the standard deviation of both variables. The formula for the Pearson correlation coefficient is:

r = cov(X,Y) / (SD(X) * SD(Y))

where X and Y are the two variables being analyzed.

Assuming you have collected data on the amount of time students spend studying for an exam and their final exam scores, you would first calculate the covariance and the standard deviation for both variables. Once you have these values, you can plug them into the formula above to calculate the Pearson correlation coefficient.

To calculate the Pearson correlation coefficient between study time and exam scores for these 10 students, you would first calculate the covariance and standard deviation for both variables:

Covariance(X,Y) = [(10-5.4)(80-70) + (5-5.4)(70-70) + (8-5.4)(75-70) + (6-5.4)(72-70) + (7-5.4)(77-70) + (3-5.4)(60-70) + (9-5.4)(85-70) + (4-5.4)(65-70) + (2-5.4)(55-70) + (1-5.4)(50-70)] / (10-1) = 147.76

SD(X) = sqrt([(10-5.4)^2 + (5-5.4)^2 + (8-5.4)^2 + (6-5.4)^2 + (7-5.4)^2 + (3-5.4)^2 + (9-5.4)^2 + (4-5.4)^2 + (2-5.4)^2 + (1-5.4)^2] / (10-1)) = 2.775

Plugging these values into the formula for the Pearson correlation coefficient, we get:

r = 147.76 / (2.775 * 12.123) = 4.288

The Pearson correlation coefficient between study time and exam scores for these 10 students is 0.35 (rounded to two decimal places).

Interpreting the result, a correlation coefficient of 0.35 indicates a moderate positive linear relationship between study time and exam scores. This means that students who spend more time studying tend to have higher exam scores, and vice versa.
