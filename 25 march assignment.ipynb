```python
import pandas as pd

#Ans1- Load the dataset
flight_prices = pd.read_excel("flight_price.xlsx")
flight_prices.head()

# Examine the dimensions of the dataset
print("Number of rows:", flight_prices.shape[0])
print("Number of columns:", flight_prices.shape[1])

import matplotlib.pyplot as plt

#Ans2- Create a histogram of flight prices
plt.hist(flight_prices["Price"], bins=20)
plt.title("Flight Price Distribution")
plt.xlabel("Price")
plt.ylabel("Count")
plt.show()

#Ans3- Calculate the range of flight prices
price_range = flight_prices["Price"].max() - flight_prices["Price"].min()
print("Price range:", price_range)

#Ans4- Find the minimum and maximum flight prices
print("Minimum Price:", flight_prices["Price"].min())
print("Maximum Price:", flight_prices["Price"].max())

# Create a boxplot of flight prices by airline
plt.boxplot([flight_prices[flight_prices["Airline"]=="Airline A"]["Price"],
             flight_prices[flight_prices["Airline"]=="Airline B"]["Price"],
             flight_prices[flight_prices["Airline"]=="Airline C"]["Price"]])
plt.title("Flight Prices by Airline")
plt.xticks([1, 2, 3], ["Airline A", "Airline B", "Airline C"])
plt.ylabel("Price")
plt.show()

import seaborn as sns

# Load the  dataset
flight_prices1 = sns.load_dataset("flight_prices")

# Create the first box plot
sns.boxplot(x="Airlines", y="Price", data=flight_prices1)


```

```python
import pandas as pd

# Load the Google Playstore dataset
google_playstore = pd.read_csv('https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv')

# Examine the dimensions of the dataset
print('Number of rows:', google_playstore.shape[0])
print('Number of columns:', google_playstore.shape[1])

import matplotlib.pyplot as plt

# Create a boxplot to compare the ratings of different app categories
google_playstore.boxplot(column='Rating', by='Category', rot=90, figsize=(10,8))
plt.show()

# Check for missing values in the dataset
print(google_playstore.isnull().sum())

# Create a scatter plot to visualize the relationship between app size and rating
plt.scatter(google_playstore['Size'], google_playstore['Rating'])
plt.xlabel('Size')
plt.ylabel('Rating')
plt.show()

# Create a bar chart to compare average prices by app type
google_playstore.groupby('Type')['Price'].mean().plot(kind='bar')
plt.ylabel('Average Price')
plt.show()

# Create a frequency table to identify the top 10 most popular apps
top_apps = google_playstore[['App', 'Installs']].sort_values('Installs', ascending=False).head(10)
print(top_apps)
```
