Q1. What is the relationship between polynomial functions and kernel functions in machine learning algorithms?

Ans1-Polynomial functions and kernel functions are both used in machine learning algorithms, but they serve different purposes. Polynomial functions are a class of functions that can be used to model complex relationships between variables. They can be used as a basis for creating features for machine learning models or as a standalone model. Kernel functions, on the other hand, are used in support vector machines (SVMs) to transform the input data into a higher dimensional space where it is easier to separate the classes. Polynomial kernels are a type of kernel function that can be used in SVMs to capture polynomial relationships between variables.

Q3. How does increasing the value of epsilon affect the number of support vectors in SVR?

Ans3-  In SVR, the epsilon parameter controls the width of the margin around the regression line where no penalty is given for errors. As the value of epsilon increases, the margin widens, and the number of support vectors may also increase. This is because larger values of epsilon allow more points to be included within the margin, which can lead to more support vectors being selected.



Q4. How does the choice of kernel function, C parameter, epsilon parameter, and gamma parameter affect the performance of Support Vector Regression (SVR)? Can you explain how each parameter works and provide examples of when you might want to increase or decrease its value?

Ans4-The choice of kernel function, C parameter, epsilon parameter, and gamma parameter can all affect the performance of SVR.

* Kernel function: The choice of kernel function determines how the input data is transformed into a higher dimensional space. Different kernel functions are suitable for different types of data and relationships between variables. For example, a linear kernel is suitable for linear relationships, while a radial basis function (RBF) kernel can capture more complex nonlinear relationships.

* C parameter: The C parameter controls the tradeoff between achieving a low training error and a low testing error. A higher value of C leads to a narrower margin and more emphasis on classifying all training data correctly, which can lead to overfitting. A lower value of C leads to a wider margin and more emphasis on achieving a low testing error, which can lead to underfitting.

* Epsilon parameter: The epsilon parameter controls the width of the margin around the regression line where no penalty is given for errors. A larger value of epsilon leads to a wider margin and more emphasis on achieving a low testing error, which can lead to underfitting. A smaller value of epsilon leads to a narrower margin and more emphasis on fitting the training data, which can lead to overfitting.

* Gamma parameter: The gamma parameter controls the width of the RBF kernel. A larger value of gamma leads to a narrower kernel and more emphasis on fitting the training data, which can lead to overfitting. A smaller value of gamma leads to a wider kernel and more emphasis on achieving a low testing error, which can lead to underfitting.

In general, it is important to tune these parameters carefully to achieve the best performance for a given problem. The optimal values will depend on the specific characteristics of the data and the goals of the analysis.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## lets make synthetic datapoints
from sklearn.datasets import make_classification
```

```python
X,y=make_classification(n_samples=1000,n_features=2,n_classes=2,n_clusters_per_class=2,n_redundant=0)
```

```python
pd.DataFrame(X)[0]
```

```python
sns.scatterplot(x=pd.DataFrame(X)[0],y=pd.DataFrame(X)[1],hue=y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.25,random_state=10)
```

```python
from sklearn.svm import SVC
```

```python
svc=SVC(kernel='linear')
```

```python
svc.fit(X_train,y_train)
```

```python
svc.coef_
```

```python
## prediction
y_pred=svc.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
```

```python
## Hyperparameter tunning with SVC
```

```python
from sklearn.model_selection import GridSearchCV

#define parameter range
param_grid={'C':[.1,1,10,100,1000],
           'gamma':[1,.1,.01,.001,.0001],
           'kernel':['linear']}
```

```python
grid=GridSearchCV(SVC(),param_grid=param_grid,refit=True,cv=5,verbose=3)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
## prediction
y_pred4=grid.predict(X_test)
print(classification_report(y_test,y_pred4))
print(confusion_matrix(y_test,y_pred4))
print(accuracy_score(y_test,y_pred4))
```
